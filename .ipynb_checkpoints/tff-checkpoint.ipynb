```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.tff.org/Default.aspx?pageID=198&hafta=9#grp")

driver.implicitly_wait(30)
team_table = driver.find_elements(by=By.CSS_SELECTOR, value=".s-table tr td:first-of-type a")

teams = list()
teams.clear() # in case this cell is re-used for whatever reason

for no in range(len(team_table)):
    name = team_table[no].text
    if no < 9:
        name = name[2:] # Deleting number from the name
    else:
        name = name[3:]
    team = {"name":name, "link":team_table[no].get_attribute("href")}
    teams.append(team)
```

```python
for team in teams:
    driver.get(team['link'])
    driver.find_element(by=By.NAME, value="ctl00$MPane$m_28_196_ctnr$m_28_196$btnAra").click()
    
    players = list()
    players.clear()
    player_table = driver.find_elements(by=By.CSS_SELECTOR, value=".GridRow_TFF_Contents td a:first-of-type")
    
    for pl in player_table:
        player = {"name":pl.text,"link":pl.get_attribute("href")}
        players.append(player)
    team.update({"players":players})
```
