```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import numpy as np
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get("https://www.transfermarkt.com/")
driver.implicitly_wait(5)

# accept the agreement
time.sleep(2)
driver.switch_to.frame(driver.find_elements(By.CSS_SELECTOR, "iframe")[-1]) # need to switch to iframe otherwise cant find it
driver.find_element(by=By.CSS_SELECTOR, value="#notice button").click()
```

```python
data = list()
data.clear() # in case this cell is re-used for whatever reason
row = dict()
row.clear()
driver.get("https://www.transfermarkt.com/super-lig/gesamtspielplan/wettbewerb/TR1")

id_no = 0
year = 1 # data will be collected from 2001 to 2022
while (year < 23):
    # I couldve done 22 to 1 but since first picked tables will be week 1 it would be weird chronologically
    time.sleep(1) # website automatically scrolls up need to wait for it
    driver.execute_script("window.scrollTo(0, 750)") # need to scroll down to click things
    driver.find_element(by=By.CSS_SELECTOR, value="table.auflistung tr:first-of-type td:nth-of-type(2) div.chzn-container > a").click()
    driver.find_element(by=By.CSS_SELECTOR, value=f"table.auflistung tr:first-of-type td:nth-of-type(2) div.chzn-container li:nth-child({24 - year})").click()
    driver.find_element(by=By.CSS_SELECTOR, value="table.auflistung tr:nth-of-type(2) td:nth-of-type(3) div.chzn-container > a").click()
    driver.find_element(by=By.CSS_SELECTOR, value="table.auflistung tr:nth-of-type(2) td:nth-of-type(3) div.chzn-container li:last-of-type").click() # so that it would choose the last week
    driver.find_element(by=By.CSS_SELECTOR, value="table.auflistung input.small.button").click()
    
    weeks = driver.find_elements(by=By.CSS_SELECTOR, value="table:not(.auflistung)") # need to do this because which week it is might also be useful
    week_no = 1
    row = {"id":None,"link":None,"year":year,"week":None,"result":None}
    for week in weeks:
        matches = week.find_elements(by=By.CSS_SELECTOR, value="td > .ergebnis-link")
        for match in matches:
            row.update({"id":id_no}) # just to make it easier to work with
            id_no += 1
            result = int(match.text[0]) - int(match.text[2])
            link = match.get_attribute("href")
            row.update({"link": link,"week":week_no, "result":result})
            data.append(row.copy())
        week_no += 1
    year += 1
```

```python
#update csv !!! don't use at restart

data = pd.DataFrame.from_dict(data)
data.to_csv("data.csv")
data
```
